Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import pylab
import seaborn as sns
from pandas import read_csv
```

Import Data

```python
csv = pd.read_csv("Index2020.csv") #record values
dataframe = csv.copy()#new variable;copy data
```

Examining the Data

```python
dataframe.isna().sum()
#spx means S&P500 Index and dax means DAX30 INDEX
```

Plotting the Data

```python
dataframe.spx.plot(figsize = (20,5), title = "S&P Prices")
plt.show()
#x-axis:the periods the values come from; indexes
#y-axis:prices
#boomin growth;falls
```

```python
dataframe.dax.plot(figsize = (20,5), title = "DAX Prices")
plt.show()
#boomin growth;sharp falls
```

```python
dataframe.spx.plot(figsize = (20,5))
dataframe.dax.plot(figsize = (20,5))
plt.title("S&P Vs. DAX")
plt.show()
```

The QQ Plot

```python
scipy.stats.probplot(dataframe.spx, plot = pylab)
pylab.show()
#y-axis:prices of market
#x-axis:how many standard deviations away from the mean
```

More Data Visualization

```python
histogram = pd.read_csv("Index2020.csv", header=0, index_col=0, parse_dates=True, squeeze=True)
histogram.hist()
plt.show()
```
