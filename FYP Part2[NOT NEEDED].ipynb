Importing the Packages

```python
import pandas as pd
import numpy as np
```

Importing the data

```python
raw_csv_data = pd.read_csv("Index2020.csv") 
df_com=raw_csv_data.copy() 
```

From Text to Date

```python
df_com.date = pd.to_datetime(df_com.date, dayfirst = True)
#method assumes string is mm/dd/yyyy form. dayfirst =true changes form to dd/mm/yyyy
```

```python
df_com.head()
```

```python
df_com.date.describe()
```

Setting the Index

```python
df_com.set_index("date", inplace=True)
```

```python
df_com.head()
```

Setting the Frequency

```python
df_com = df_com.asfreq('d')
```

```python
df_com
```

```python
df_com = df_com.asfreq('b')
```

```python
df_com.head()
```

Handling Missing Values

```python
df_com.isna()
```

```python
df_com.isna().sum()
```

```python
df_com.spx = df_com.spx.fillna(method='ffill')
#frontfill
```

```python
df_com.dax = df_com.dax.fillna(method='ffill')
```

```python
df_com.isna().sum()
```

Simplifying the Dataset

```python
del df_com['dax']
```

```python
df_com.describe()
```
